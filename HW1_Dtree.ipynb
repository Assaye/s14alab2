```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./data/Boston_March2018.csv')
```

```python
y = data['SOLDPRICE']
X = data[['BEDS', 'BATHS', 'SQFT', 'AGE', 'GARAGE']]
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python




# Split the data into training/testing sets @ 33% size (sklearn)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

```python
dt = DecisionTreeRegressor(max_depth=4, min_samples_leaf=0.1,random_state=3)
```

```python
dt.fit(X_train, y_train)
```

```python
y_pred = dt.predict(X_test)
```

```python
mse_dt = mean_squared_error(y_test, y_pred)
```

```python
rmse_dt = mse_dt**(1/2)
```

```python
print('Mean squared error: %.2f' % rmse_dt)
```

```python
print('Variance score: %.2f' % r2_score(y_test, y_pred))
```

```python
# Plot outputs
plt.figure(figsize=(20, 10))
plt.scatter(y_test, y_pred, color='black')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=4, color='blue')

plt.ylabel('Predicted')
plt.xlabel('Measured')

plt.xticks(())
plt.yticks(())

plt.show()
```
